```python
import pandas as pd
import numpy as np
from functools import reduce
from collections import Counter
import os
import time
from IPython.display import display, clear_output
```

```python
file_name = '반려동물'
data = pd.read_csv('./'+file_name+'_result.csv')
```

```python
data = data.sort_values(by = 'label')
```

```python
category = []
stop_word = ['그리고','이렇게','제발','또한','입니다','너무','다','정말','하고','하는',"합니다",'이런','있는','저는','있습니다','겁니다','빨리','주세요','또','한','및','등','할','그런데','것으로','위해','이','그','수','했더니','하니','해서','저희','왜','큰','그','할','안','인해','그래서','살고','대한','다른','많이','많은','제가','있다','건은','있습니다','빨리','한다','이번','건은','모든','없습니다','아니라','작은','위해','이','주십시요','아무것도,','위한','진짜'
            ,'아니','않고','이에','등','등의','최근','통해','대해','해당','현재','만큼','하게','단지','보면','봅니다','해야','것을','아닌'
            ,'같은','제대로','어떻게','생각합니다','안녕하세요','보고','하지','지금','없는','것은','것이']
for i in range(250):
    data_tmp = data[data['label'] == i]
    texts = list(data_tmp['text'])
    words = list(map(lambda x:x.split(' ') , texts))
    words = list(reduce(lambda x, y: x+y, words))
    words = list(filter(lambda x:len(x) > 1 , words))
    
    counter = Counter(words)
    print(counter)
    for sw in stop_word:
        counter[sw] = 0
    
    clear_output(wait=True)
    display(sorted(list(counter.items()) ,key = lambda x:x[1])[-20:])
    
    
    
    time.sleep(1)
    
    cate = input()
    category.append([i,cate])
    
data['category_second'] = ['' for i in range(len(data))]
tmp = []
for i in range(len(data)):
    tmp.append(category[data.iloc[i]['label']][1])
data['category_second'] = tmp

data.to_csv('./'+file_name+'라벨링.csv')
    
```
